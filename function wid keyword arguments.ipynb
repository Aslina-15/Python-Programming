```python
def student_report(name,grade,subject="math",mark=0):
    print("--------student_report---------")
    print(f"name   :{name}")
    print(f"grade   :{grade}")
    print(f"subject  :{subject}")
    print(f"mark   :{mark}")
    if mark>=50:
        print("status pass")
    else:
        print("status fail")
    print("--------------------------------\n")

student_report(name="alice",grade="7th",subject="science",mark=78)
student_report(name="charlie",grade="6th",mark=45)
student_report(name="job",grade="8th")

    
```

```python
def calculate_bill(amount,tax_rate=0.05,discount=0):
    tax=amount*tax_rate
    total=amount+discount
    print(f"amount :₹{amount}")
    print(f"tax  :₹{tax}")
    print(f"discount  :₹{discount}")
    print(f"total bill  :₹{total}\n")
    
calculate_bill(amount=100, discount=50)
calculate_bill(tax_rate=0.1, amount=500)
calculate_bill(amount=1500)
```

```python
def bmi(weight,height, unit="metric"):
 if unit == "metric":
    bmi_value = weight/(height**2)
 else:
    bmi_value = 703*weight/(height**2)
     
print(f"bmi({unit})=({bmi_value: 2f}")
bmi(weight=70,height=1.75)
bmi(weight=154,height=68,unit="imperial")

```
