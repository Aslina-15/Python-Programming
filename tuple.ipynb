```python
# to change the elements of tuple 
a=(4,5,6,7)
a=a[:1]+(a,)+a[2:]
print(a)
```

```python
#swaping of number in tuple 
x=7
y=8
x,y=y,x
print(x,y)
```

```python
#count number of elements in tuple
t=(1,2,3,4,5,6,7,2,2)
c=t.count(2)
print(c)
```

```python
#tuple unpacking 
data=(110,"Aslina",18)
rollno,name,age=data
print(rollno,name,age)
```

```python
#list and tuple data type transformation
l=[1,2,3]
t=[4,5,6]
print(l,t)
print(list(t),tuple(l))
```

```python
#for loop in tuple
sports=("football","throwball","volleyball","basketball")
for i in sports:
    print(i)
for i in range(len(sports)):
    print(sports[i])
```

```python
#list of tuples
student=[("ram",20),("ravi",18),("arjun",19)]
print(student)
sort=sorted(student,key=lambda x:x[1])
print(sort)
sort1=sorted(student,key=lambda x:x[0])
print(sort1)
```

```python
#tuple string
a=(2,4,6,7,8)
print(a[::-1])
print(a[0::2])
print(a[-5:])
print(a[::2])
print(a[0:])
```
