```python
#arithemetic operation with default arguments 
def calculate(a,b, operation="add"):
 if operation == "add":
    return a+b
 elif operation == "sub":
    return a-b
 elif operation == "mul":
    return a*b
 elif operation == "div":
    return a/b
 elif operation == "mod":
    return a%b
 else:
    print("unknown operation")
print(calculate(5,2))
print(calculate(5,2,"sub"))
print(calculate(5,2,"mul"))
print(calculate(5,2,"div"))
print(calculate(5,2,"mod"))





```

```python
# calculate total price with default tax 
def totalprice(price,quantity=1,tax_rate=0.05):
    total=price*quantity
    tax=total*tax_rate
    return total+tax
print("total default quantity and tax:",totalprice(200))
print("total custom quantity and default tax:",totalprice(300,2))
print("total all values provided:",totalprice(400,3,0.09))
```

```python
#salary calculation
def salary(basic,hra=0.2,da=0.1):
    hra_amount=basic*hra
    da_amount=basic*da
    total_salary=basic+hra_amount+da_amount
    return total_salary
print("salary with default hra and da:",salary(30000))
print("salary with custom hra and default da:",salary(50000,0.3))
print("salary with custom hra and da:",salary(60000,0.4,0.2))
```

```python
#using lambda with map function
n=[4,6,8,9,2]
square=list(map(lambda n:n*n,n))
print(square)
n=[4,6,1,3,8,9,2]
e=list(map(lambda n:n%2==0,n))
print(e)
```

```python
#using filter()
n=[4,6,1,3,8,9,2]
e=list(filter(lambda n:n%2==0,n))
print(e)
```
