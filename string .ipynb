```python
#count uppercase,lowercaseand digits in a sentence
s=input("enter the sentence:")
u=0
l=0
d=0
for i in s:
    if i.isupper():
        u=u+1
    elif i.islower():
        l=l+1
    elif i.isdigit():
        d=d+1
print(" no of uppercase:",u)
print(" no of lowercase:",l)
print(" no of didgit:",d)
    
```
