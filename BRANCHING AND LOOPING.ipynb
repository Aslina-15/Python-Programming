```python
# check whether a number is odd or even:

n=int(input("enter a number:"))
if n%2==0:
    print("even")
else:
    print("odd")
```

```python
# leap year checker using if else copndition :
year=int(input("enter the year:"))
if(year%4==0 and year%100!=0):
  print("its a leap year")
else:
  print("its not a leap year")
```

```python
# to calculate the student grade details:

rollno=int(input("enter your rollno:"))
name=input("enter your name:")
m1=int(input("enter m1:"))
m2=int(input("enter m2:"))
m3=int(input("enter m3:"))
m4=int(input("enter m4:"))
m5=int(input("enter m5:"))
total=m1+m2+m3+m4+m5
print("total marks:",total)
avg=total/5
print("avgerage:",avg)
if avg>=90 and avg<100:
    print ("A+ grade")
elif avg>=80 and avg<90:
    print ("A grade")
elif avg>=70 and avg<80:
    print ("B grade")
elif avg>=60 and avg<70:
    print ("C grade")
elif avg>=50 and avg<60:
    print ("D grade")
elif avg>=40 and avg<50:
    print ("E grade")
elif avg>=35 and avg<40:
    print ("Just Pass")
else:
    print("Fail")
    
```

```python
# reverse a number using while loop
n=int(input("enter a number:"))
s=0
while n!=0:
   r=n%10
   s=s*10+r
   n=n//10
print("reverse of a number:",s)
```

```python
# factorial of number using while loop 

n=int(input("enter a number:"))
a=1
fact=1
while a<=n:
     fact*=a
     a+=1
print("factorial of a number is:",fact)
```
