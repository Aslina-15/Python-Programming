```python
#factorial of number

def factorial(n):
       if n==0:
         return 1 
       else:
         return n*factorial(n-1)
num=int(input("enter a number:"))
if num<0:
          print("factorial does not exist for -ve number:")
else:
          print(f"the factorial of{num}is{factorial(num)}")
        
```

```python
# fibanacci series (n th term)
def fibonacci(n):
    if n<=1:
        return n
    else:
        return fibonacci(n-1)+fibonacci(n-2)
num=int(input("enter position (n) in fibonacci series:"))
if num<0:
    print("fibonacci number does not exist for -ve index:")
else:
    print(f"the {num} the fibonacci number is {fibonacci(num)}")
```
