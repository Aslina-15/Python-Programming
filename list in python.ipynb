```python
#creating a list
numbers=[2,4,6,8,10]
colors=["blue","pink","red","white","black"]
combined_list=[110,80.9,"aslina",True]
empty_list=[]
print(numbers,colors,combined_list,empty_list)
```

```python
#accessing the list elements 
colors=["red","pink","blue","white","black"]
print(colors[0])
print(colors[-5])
print(colors[-1])
print(colors[4])
```

```python
#modify list
colors=["red","pink","blue","white","black"]
colors[1]="lavender"
colors[2]="green"
print(colors)
```

```python
#common list operation 
colors=["red","pink","blue","white","black"]
colors.pop()
print(colors)
```

```python
#common list operation 2
colors=["red","pink","blue","white","black"]
colors.remove("blue")
print(colors)
```

```python
#common list operation 3
colors=["red","pink","blue","white","black"]
combined_list=[110,80.9,"Aslina",True]
print(len(colors))
print(len(combined_list))
```

```python
#Buildin function in a list
a=[5,34,98,20,25]
print(min(a))
print(max(a))
a.sort()
print(a)
a.reverse()
print(a)
```

```python
#looping in a list 
colors=["red","pink","blue","white","black"]
for i in colors:
    print(i)
```

```python
#list sorting 
a=[5,34,98,20,25]
print(a[0:2])
print(a[::-1])
```

```python
#square of number from ranges 1 to 5
s=[n**2 for n in range(1,6)]
print(s)
```

```python
#square of number 
l1=[5,2,4,7,8]
s=[n**2 for n in l1]
print(s)
```

```python
#count how many number are even in list
z=[3,8,9,10,45,22]
w=0
for n in z:
    if n%2==0:
       w+=1
    print(w)
    
```
