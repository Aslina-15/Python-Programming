```python
#creating a set
color={"red","black","white"}
print(color)
```

```python
#creating a set using set function 
n=set([2,1,3,2])
print(n)
```

```python
#basic set operation
a={1,2,3}
b={3,4,5}
print(a.union(b))
print(b.union(a))

print(a.intersection(b))
print(b.intersection(a))

print(a.difference(b))
print(b.difference(a))

print(a.symmetric_difference(b))
print(b.symmetric_difference(a))
```

```python
#adding new element in set
os={"windows","unix","linux"}
os.add("mac os")
print(os)
```

```python
#subject chosen
maths={"aslina","asma","sakthi","harini","krishna"}
cs={"aslina","krishna","sree","nithin","mouli"}
print("students who are chosen both subject:",maths.intersection(cs))
print("students who are chosen either maths (or) cs :",maths.symmetric_difference(cs))
print("students who are chosen maths alone:",maths.difference(cs))
print("students who are chosen cs alone:",cs.difference(maths))
maths.add("harthika")
print(maths)
```

```python
#updating a set 
maths={"aslina","asma","sakthi","harini","krishna"}
m={"dhanu","subha"}
maths.update(m)
print(maths)
maths.update({"roshan","fahd","surya"})
print(maths)
```

```python
#type error due to unhasable set elements:
maths={"aslina","asma","sakthi","harini","krishna"}
maths.add({"roshan","fahd","surya"})
print(maths)
```

```python
#discard function
maths={"aslina","asma","sakthi","harini","krishna"}
maths.discard("aslina")
print(maths)
```

```python
#remove function
maths={"aslina","asma","sakthi","harini","krishna"}
maths.remove("sakthi")
print(maths)
```

```python
#avoid duplicates
s={'a','c','e','d'}
s.update("and")
print(s)
```

```python
#clear function
s={'a','c','e','d'}
s.clear()
print(s)
```

```python
#key error while removing non-existed element from set 
maths={"aslina","asma","sakthi","harini","krishna"}
maths.discard("aslina")
maths.remove("aslina")
print(maths)
```
